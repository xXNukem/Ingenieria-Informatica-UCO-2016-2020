```python
import numpy as np
import matplotlib.pyplot as mp
import pandas as pd
from scipy.io import arff
from sklearn.preprocessing import MinMaxScaler

dataD=arff.loadarff('./datasets/diabetes.arff')
dataI=arff.loadarff('./datasets/iris.arff')
dataW=arff.loadarff('./datasets/glass.arff')


dfD = pd.DataFrame(dataD[0])
dfD.plot.hist(bins=12, alpha=0.5, by=4)
mp.suptitle('DIABETES')


dfI = pd.DataFrame(dataI[0])
dfI.plot.hist(bins=12, alpha=0.5, by=4)
mp.suptitle('IRIS')


dfW = pd.DataFrame(dataW[0])
dfW.plot.hist(bins=12, alpha=0.5, by=4)
mp.suptitle('GLASS')

mp.show()
```

<b>Respuesta: En los histogramas de atributos se pueden ver representada la frecuencia con la que dichos atributos aparecenen el dataset. Cada uno se representa de forma intuitiva con un color distinto, esto es algo que el sistema hace internamente y resulta muy útil a la hora de analizar el histograma. Además de representar la frecuencia de aparición de cada atributo en el eje Y, en el eje X se muestra el rango de valores perteneciente a cada frecuencia.

<b>Ejercicio 3: Estudie el efecto de la normalización (reescalar en el intervalo [0, 1]) y la estandarización (μ = 0, σ = 1) sobre
el histograma

```python
import numpy as np
import matplotlib.pyplot as mp
import pandas as pd
from scipy.io import arff
from sklearn.preprocessing import MinMaxScaler

dataD=arff.loadarff('./datasets/diabetes.arff')
dataI=arff.loadarff('./datasets/iris.arff')
dataW=arff.loadarff('./datasets/glass.arff')

scaler=MinMaxScaler()


dfD = pd.DataFrame(dataD[0])
scaled_df=scaler.fit_transform(dfD.iloc[:, :-1])
scaled_df = pd.DataFrame(scaled_df)
scaled_df.plot.hist(bins=12, alpha=0.5, by=4)
mp.suptitle('DIABETES NORMALIZADO')


dfI = pd.DataFrame(dataI[0])
scaled_df=scaler.fit_transform(dfI.iloc[:,:-1])
scaled_df= pd.DataFrame(scaled_df)
scaled_df.plot.hist(bins=12, alpha=0.5, by=4)
mp.suptitle('IRIS NORMALIZADO')


dfW = pd.DataFrame(dataW[0])
scaled_df=scaler.fit_transform(dfW.iloc[:,:-1])
scaled_df=pd.DataFrame(scaled_df)
scaled_df.plot.hist(bins=12, alpha=0.5, by=4)
mp.suptitle('GLASS NORMALIZADO')

#-----------------------------------------------------------------------

dfD = pd.DataFrame(dataD[0])
est=dfD.iloc[:,:-1]
est_df=(est - np.mean(est))/np.std(est)
est_df.plot.hist(bins=12, alpha=0.5, by=4)
mp.suptitle('DIABETES ESTANDARIZADO')


dfI = pd.DataFrame(dataI[0])
est=dfI.iloc[:,:-1]
est_df=(est - np.mean(est))/np.std(est)
est_df.plot.hist(bins=12, alpha=0.5, by=4)
mp.suptitle('IRIS ESTANDARIZADO')


dfW = pd.DataFrame(dataW[0])
est=dfW.iloc[:,:-1]
est_df=(est - np.mean(est))/np.std(est)
est_df.plot.hist(bins=12, alpha=0.5, by=4)
mp.suptitle('GLASS ESTANDARIZADO')

mp.show()
```

<b> Los métodos de estandarizacion y normalización han ajustado los tres conjuntos de datos dispuestos en diferentes escalas a una escala comun. La estandarizacion pone todos los valores de los atributos entre 0 y 1. La estandarizción hará que los datos sigan una distribución normal, ya que como se puede ver en la ejecución del script, los histogramas cogen una forma similar a la de una campana de GAUSS.
