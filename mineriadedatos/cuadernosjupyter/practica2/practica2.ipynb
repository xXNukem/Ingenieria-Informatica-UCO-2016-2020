<b>Ejercicio 3:</b></br> Estudie el efecto de la normalización (reescalar en el intervalo [0, 1]) y la estandarización (μ = 0, σ = 1) sobre
el histograma

```python
import numpy as np
import matplotlib.pyplot as mp
import pandas as pd
from scipy.io import arff
from sklearn.preprocessing import MinMaxScaler

dataD=arff.loadarff('./datasets/diabetes.arff')
dataI=arff.loadarff('./datasets/iris.arff')
dataW=arff.loadarff('./datasets/glass.arff')

scaler=MinMaxScaler()


dfD = pd.DataFrame(dataD[0])
scaled_df=scaler.fit_transform(dfD.iloc[:, :-1])
scaled_df = pd.DataFrame(scaled_df)
scaled_df.plot.hist(bins=12, alpha=0.5, by=4)
mp.suptitle('DIABETES NORMALIZADO')


dfI = pd.DataFrame(dataI[0])
scaled_df=scaler.fit_transform(dfI.iloc[:,:-1])
scaled_df= pd.DataFrame(scaled_df)
scaled_df.plot.hist(bins=12, alpha=0.5, by=4)
mp.suptitle('IRIS NORMALIZADO')


dfW = pd.DataFrame(dataW[0])
scaled_df=scaler.fit_transform(dfW.iloc[:,:-1])
scaled_df=pd.DataFrame(scaled_df)
scaled_df.plot.hist(bins=12, alpha=0.5, by=4)
mp.suptitle('GLASS NORMALIZADO')

#-----------------------------------------------------------------------

dfD = pd.DataFrame(dataD[0])
est=dfD.iloc[:,:-1]
est_df=(est - np.mean(est))/np.std(est)
est_df.plot.hist(bins=12, alpha=0.5, by=4)
mp.suptitle('DIABETES ESTANDARIZADO')


dfI = pd.DataFrame(dataI[0])
est=dfI.iloc[:,:-1]
est_df=(est - np.mean(est))/np.std(est)
est_df.plot.hist(bins=12, alpha=0.5, by=4)
mp.suptitle('IRIS ESTANDARIZADO')


dfW = pd.DataFrame(dataW[0])
est=dfW.iloc[:,:-1]
est_df=(est - np.mean(est))/np.std(est)
est_df.plot.hist(bins=12, alpha=0.5, by=4)
mp.suptitle('GLASS ESTANDARIZADO')

mp.show()
```

<b>Respuesta: </b></br> Los métodos de estandarizacion y normalización han ajustado los tres conjuntos de datos dispuestos en diferentes escalas a una escala comun. La normalización pone todos los valores de los atributos entre 0 y 1. La estandarizción hará que los datos sigan una distribución normal, ya que como se puede ver en la ejecución del script, los histogramas cogen una forma similar a la de una campana de GAUSS.

<b>Ejercicio 8: </b> </br>Estudie el diagrama de correlaciones de los tres conjuntos e indique qué información relativa a las diferentes clases puede obtener.

```python
import numpy as np
import matplotlib.pyplot as mp
import pandas as pd
from scipy.io import arff
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from pandas.plotting import scatter_matrix

dataD=arff.loadarff('./datasets/diabetes.arff')
dataI=arff.loadarff('./datasets/iris.arff')
dataW=arff.loadarff('./datasets/glass.arff')

dfD = pd.DataFrame(dataD[0])
plt.matshow(dfD.corr())
plt.xticks(range(len(dfD.columns)), dfD.columns)
plt.yticks(range(len(dfD.columns)), dfD.columns)
plt.colorbar()

dfI = pd.DataFrame(dataI[0])
plt.matshow(dfI.corr())
plt.xticks(range(len(dfI.columns)), dfI.columns)
plt.yticks(range(len(dfI.columns)), dfI.columns)
plt.colorbar()

dfW = pd.DataFrame(dataW[0])
plt.matshow(dfW.corr())
plt.xticks(range(len(dfW.columns)), dfW.columns)
plt.yticks(range(len(dfW.columns)), dfW.columns)
plt.colorbar()

mp.show()
```

<b>Respuesta: </b> </br>El diagrama de correlación nos permite observar la proporcionalidad y la relacion que existe entre las variables que componen el conjunto de datos. Esto nos da a entender que, si cuando una variable se modifica también lo hace otra, estas dos variables estarán correlacionadas en mayor o menor grado. Por ejemplo en el dataset iris se observa que las variables petalenght y petalwidth están muy correlacionadas
